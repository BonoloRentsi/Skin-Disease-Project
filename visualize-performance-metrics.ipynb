```python
data = {
    'Model': ['XCEPTIONNET'] * 10 + ['DENSENET201'] * 10 + ['RESNET152V2'] * 10 + ['CNN'] * 10,
    'Epoch': list(range(1, 11)) * 4,
    'Accuracy': [
        0.8665, 0.9092, 0.9095, 0.9032, 0.9131, 0.9090, 0.9077, 0.9075, 0.9232, 0.9271,
        0.8364, 0.9315, 0.9353, 0.9395, 0.9410, 0.9419, 0.9420, 0.9418, 0.9422, 0.9443,
        0.8963, 0.9373, 0.9367, 0.9387, 0.9473, 0.9461, 0.9505, 0.9523, 0.9530, 0.9594,
        0.8576, 0.9040, 0.9080, 0.9160, 0.9139, 0.9140, 0.9170, 0.9224, 0.9219, 0.9246
    ]
}

df = pd.DataFrame(data)

sns.set(style="whitegrid")

plt.figure(figsize=(18, 12))
sns.lineplot(data=df, x='Epoch', y='Accuracy', hue='Model', markers=True, dashes=False)

plt.title('MODEL ACCURACY ACROSS EPOCHS', fontsize=16)
plt.xlabel('EPOCH', fontsize=12)
plt.ylabel('ACCURACY', fontsize=12)


plt.ylim(0.80, 1.00) 
plt.xticks(range(1, 11))
plt.grid(True)
plt.savefig('model_accuracy_epochs1111.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

epochs = list(range(1, 11))
xceptionnet_accuracy = [0.8665, 0.9092, 0.9095, 0.9032, 0.9131, 0.9090, 0.9077, 0.9075, 0.9232, 0.9271]
densenet_accuracy = [0.8364, 0.9315, 0.9353, 0.9395, 0.9410, 0.9419, 0.9420, 0.9418, 0.9422, 0.9443]
resnet_accuracy = [0.8963, 0.9373, 0.9367, 0.9387, 0.9473, 0.9461, 0.9505, 0.9523, 0.9530, 0.9594]
cnn_accuracy = [0.8576, 0.9040, 0.9080, 0.9160, 0.9139, 0.9140, 0.9170, 0.9224, 0.9219, 0.9246]
plt.figure(figsize=(10, 6))
plt.plot(epochs, xceptionnet_accuracy, label='XceptionNet', marker='o', color='red')
plt.plot(epochs, densenet_accuracy, label='DenseNet201', marker='o', color='green')
plt.plot(epochs, resnet_accuracy, label='ResNet152V2', marker='o', color='blue')
plt.plot(epochs, cnn_accuracy, label='CNN', marker='o', color='orange')
plt.title('Model Training Accuracy Across Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Updated data to include 'AUC-ROC'
data = {
    'Model': ['XceptionNet', 'XceptionNet', 'XceptionNet', 'XceptionNet',
              'ResNet152V2', 'ResNet152V2', 'ResNet152V2', 'ResNet152V2',
              'CNN', 'CNN', 'CNN', 'CNN',
              'DenseNet201', 'DenseNet201', 'DenseNet201', 'DenseNet201'],
    'Metric': ['Sensitivity', 'Specificity', 'Precision', 'F1-Score'] * 4,
    'Value': [
        1.00, 0.99, 0.95, 0.93, 
        1.00, 1.00, 0.96, 0.96,  
        0.94, 0.94, 0.95, 0.94,  
        1.00, 0.98, 0.95, 0.94   
    ]
}

# Add the AUC-ROC metric
auc_data = {
    'Model': ['XceptionNet', 'ResNet152V2', 'CNN', 'DenseNet201'],
    'Metric': ['AUC-ROC', 'AUC-ROC', 'AUC-ROC', 'AUC-ROC'],
    'Value': [0.99, 1.0, 1.0, 1.0]
}

df_metrics = pd.DataFrame(data)
df_auc = pd.DataFrame(auc_data)
df = pd.concat([df_metrics, df_auc], ignore_index=True)
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
bright_palette = ['#004d99', '#88B04B', '#FFBD33', '#FF5733','#ff6666'] 

bar_plot = sns.barplot(data=df, y='Model', x='Value', hue='Metric', palette=bright_palette)
plt.title('MODEL PERFORMANCE METRICS', fontsize=16)
plt.xlabel('Score', fontsize=12)
plt.ylabel('Model', fontsize=12)

plt.legend(title='METRIC', loc='upper left', bbox_to_anchor=(1, 1), fontsize='10')
plt.xlim(0, 1.05)
plt.grid(axis='x')
plt.tight_layout() 
plt.savefig('model_metrics.png')
plt.show()
```
